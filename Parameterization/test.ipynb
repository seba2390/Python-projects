```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def circle_spiral(t):
    """
    Args:
        t = np.array([t_1,t_2,...,t_n])
    Return:
        x,y = np.array([x_1,x_2,...,x_n]), np.array([y_1,y_2,...,y_n])
    """
    return t * np.cos(t) / (2 * np.pi),  t * np.sin(t) / (2 * np.pi)
    
```

```python
t = np.linspace(0,30,1000)
x,y = circle_spiral(t)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.plot(x,y)
ax.set_aspect("equal")
ax.grid()
plt.show()
```
