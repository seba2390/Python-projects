# Spredning på delta-potentiale
I denne notebook simulerer vi en fri partikel  der spredes på et potential på formen $V(x) = -\alpha\delta(x)$. Gennem hele opgaven arbejder vi i enheder hvor $\hbar =m= 1$.

Vi starter ud med en gaussisk bølgepakke på formen 
$$
\Psi(0,x) \propto \exp \left[ -\left(\frac{x-x_0}{a}\right)^2\right]\exp(ik_0x).
$$
Her er $x_0$ middelværdien for positionen af vores partikel, $a$ er et udtryk for bredden af vores startbølgefunktion og $k_0$ er middelimpulsen (idet $\hbar = 1$). Vi vil gerne have at partiklen starter langt til venstre og bevæger sig mod højre, så fornuftige værdier af $x_0, a $ og $k_0$ opfylder $x_0\ll -a$ og $k_0 \gg 1/a$.

I kodecellen nedenfor importerer vi de nødvendige biblioteker og plotter start bølgefunktionen.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
# Indstillinger for psi0
a = 3 # Når a bliver større bliver funktionen bredere
alpha = 3 # Styrken af delta-funktionspotentialet
k0 = 2 # Middelimpuls for bølgefunktionen
x0 = -10 # Start-middelposition for bølgefunktionen

# Vi vil benytte et interval på x-aksen med bredde 2*L omkring deltafunktionen, hvor L er "stor nok":
L=5*np.abs(x0)
# Definer x punkter.
x_arr = np.linspace(-L, L, 1000)

# Psi_0 har form som en gaussisk funktion
psi0 = np.exp(-((x_arr-x0)/a)**2, dtype=complex) 
psi0 *= np.exp(1j*k0*x_arr) # Gang med exponential faktor
psi0 /= np.sqrt(np.trapz(np.abs(psi0)**2, x_arr)) # og normer

print(np.trapz(np.abs(psi0)**2, x_arr))

# Nedenfor plottes bølgefunktionen
fig, axs = plt.subplots(1, 2, figsize=(18, 6))  # Lav en figur med 2 axer
# Plot real og imaginær del på første akse
axs[0].plot(x_arr, np.real(psi0), color='r', label=r'$Re(\Psi)$')
axs[0].plot(x_arr, np.imag(psi0), color='b', label=r'$Im(\Psi)$')
xlim = (-2*np.abs(x0), 2*np.abs(x0))
axs[0].set_xlim(*xlim)
axs[0].set_xlabel('$x$')
axs[0].set_ylabel(r'$\Psi(x,0)$')
axs[0].legend()
axs[1].plot(x_arr, np.abs(psi0)**2, color='k', label=r'$|\psi|^2$')
axs[1].set_xlim(*xlim)
axs[1].set_xlabel('$x$')
axs[1].set_ylabel(r'$|\Psi(x,0)|^2$')
plt.show()
```

Vi vil nu beregne vores startbølgefunktion som en superposition af stationære tilstande. De stationære tilstande er givet i lærebogens ligninger (2.134) og (2.135), hvor vi sætter $A = 1$, $G = 0$ og lader de $k$-afhængige $B$ og $F$ være defineret som i ligning (2.140). De stationære tilstande bliver så:
$$
\psi_k(x) = \Theta(-x)\left(e^{ikx}+\frac{i\beta}{1-i\beta}e^{-ikx}\right)+\Theta(x) \frac{1}{1-i\beta} e^{ikx}
$$
hvor $\beta = \alpha/k$ (i vores enheder med $\hbar = m = 1$) og
$$
\Theta = 
\begin{cases}
1 \quad \text{hvis } x>0\\
0 \quad \text{hvis } x<0
\end{cases}
$$
er Heavisides step funktion. Bemærk, ovenstående svarer også det, der udledes i videoen om spredningstilstande i delta-funktionspotentialet (ca. 14 minutter inde i videoen).

Vi gør nu som i tilfældet med den fri partikel og skriver:
$$
\Psi(x,t) = \frac{1}{\sqrt{2\pi}}\int_0^{\infty} \phi(k) \psi_k(x) e^{-i\frac{\hbar k^2}{2m}t} dk,
$$
hvilket ligner ligning (2.101) rigtigt meget, men med integration fra 0 til $\infty$ og med ovenstående $\psi_k(x)$ i stedet for $e^{ikx}$. Til tiden $t=0$ har vi:
$$
\Psi(x,0) = \frac{1}{\sqrt{2\pi}}\int_0^{\infty} \phi(k) \psi_k(x) dk \approx \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} \phi(k) e^{ikx} dk,
$$
siden vi for en passende lokaliseret startbølgefunktion (godt til venstre for $x=0$) ikke kan se forskel på løsningen til deltafunktionspotentialet og til den fri partikel. Vi kan også blot antage at $\phi(k) = 0$ for $k < 0$ efter andet lighedstegn ovenfor. Vi kan derfor benytte ligning (2.104) til at finde $\phi(k)$.

I praksis skal vi definere en endelig vektor af $k$-værdier, og et pragmatisk bud er at benytte $k=\frac{\pi}{2L}n$, hvor $n$ er naturlige tal.

```python
# Det antal af k-værdier vi ønsker at bruge
n_max = 500

# Nedenfor laves et array af k'er. k=0 skal ikke være med da psi_k=0 for k=0 
dk = np.pi/2/L
k_arr = (np.arange(n_max)+1)*dk


# Array til at gemme phi_k
phi_k = [] 

# Arrays til at gemme R_k og T_k:
R_k = [];
T_k = [];

# Vi gemmer de stationære tilstande så vi let kan bruge dem senere
psi_k_mat = [] 

# I det følgende loop regner vi de stationære tilstande. Vi regner desuden
# phi_k, R_k og T_k:
for k in k_arr:
    B_k = 1j*alpha/(k-1j*alpha)
    F_k = k/(k-1j*alpha)
    psi_k = np.heaviside(-x_arr,.5)*(np.exp(1j*x_arr*k)+B_k*np.exp(-1j*x_arr*k))\
            + np.heaviside(x_arr,.5)*F_k*np.exp(1j*x_arr*k)
    psi_k_mat.append(psi_k)
    phi_k.append(np.trapz(psi0 * psi_k.conj(), x=x_arr)/np.sqrt(2*np.pi))  # Dette er ligning (2.104)
    R_k.append(np.abs(B_k)**2) # Dette er ligning (2.141) med A = 1.
    T_k.append(np.abs(F_k)**2) # Dette er ligning (2.142) med A = 1.

# Lav phi(k) om til et numpy array:
phi_k = np.array(phi_k) 

# Lav matrix med de stationære tilstande som søjler
psi_k_mat = np.array(psi_k_mat).T 

# Definer en vektor c_k, som blot er en skaleret udgave af phi_k (se forklaring nedenfor)
c_k = phi_k*dk/np.sqrt(2*np.pi)

# Vi plotter også lige phi(k)-funktionen:
fig, axs = plt.subplots(1, 2, figsize=(18, 6))  # Lav en figur med 2 axer
# Plot real og imaginær del på første akse
axs[0].plot(k_arr, np.real(phi_k), color='r', label=r'$Re(\phi)$')
axs[0].plot(k_arr, np.imag(phi_k), color='b', label=r'$Im(\phi)$')
axs[0].set_xlabel('$k$')
axs[0].set_ylabel(r'$\phi(k)$')
axs[0].legend()
# Plot norm-kvadrat i den anden figur. Vi plotter også lige de k-afhængige R og T koefficienter.
axs[1].plot(k_arr, np.abs(phi_k)**2, color='k', label=r'$|\phi|^2$')
axs[1].plot(k_arr, R_k, color='g', label=r'$R_k$')
axs[1].plot(k_arr, T_k, color='m', label=r'$T_k$')
axs[1].set_xlabel('$k$')
axs[1].set_ylabel(r'$|\phi(k)|^2$')
axs[1].legend()
plt.show()
```

Nu hvor vi kender $\phi(k)$, er vi klar til at beregne den tidsafhængige bølgefunktion, som bliver (her med $\hbar = m = 1$):
$$
\Psi(x,t) = \frac{1}{\sqrt{2\pi}}\int_0^{\infty}\phi(k)\psi_k(x) e^{-i\frac{k^2}{2} t} dk \approx  \sum_k \left\{ \frac{\phi(k) dk}{\sqrt{2\pi}}\right\} \psi_k(x) e^{-i\frac{k^2}{2} t} \equiv \sum_k c_k \psi_k(x) e^{-i\frac{k^2}{2}t},
$$
hvor approksimationen svarer til at skrive integralet som en Riemann-sum (med $dk$ som defineret i koden ovenfor) og sidste lighedstegn definerer $c_k$ som tallet i Tuborg-parenteserne.

Nedenfor skriver vi en update funktion, der opdaterer koefficienterne $c_k\rightarrow c_k e^{-i\frac{k^2}{2}t}$, og regner summen ovenfor.

```python
# De tider hvor bølgefunktionen bestemmes
t_arr = np.linspace(0, 4*np.abs(x0)/k0, 50)

# Gør figuren klar
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
re_line, = axs[0].plot([], [], color='r')
im_line, = axs[0].plot([], [], color='b')
axs[0].set_xlim(-L, L)
axs[0].set_ylim(-np.max(np.abs(psi0))*1.5, np.max(np.abs(psi0))*1.5)
axs[0].set_xlabel('$x$')
axs[0].set_ylabel(r'$\Psi(x,t)$')
abs_line, = axs[1].plot([], [], color='k')
axs[1].set_xlim(-L, L)
axs[1].set_ylim(-np.max(np.abs(psi0)**2)*1.5, np.max(np.abs(psi0)**2)*1.5)
axs[1].set_xlabel('$x$')
axs[1].set_ylabel(r'$|\Psi(x,t)|^2$')


# Dette er funktionen der opdaterer figuren hver frame.
def update(i):
    # t for den pågældende frame
    t = t_arr[i] 
    
    # Opdater coeffi
    coeffs = c_k * np.exp(-1j*k_arr**2/2 * t)
    
    # Regn bølgefunktionen til t og skaler
    psi_t = psi_k_mat @ coeffs
    psi_t = np.array(psi_t, dtype=complex)
    
    # Opdater artists for figur
    re_line.set_data(x_arr, np.real(psi_t))
    im_line.set_data(x_arr, np.imag(psi_t))
    abs_line.set_data(x_arr, np.abs(psi_t)**2)
    return re_line, im_line, abs_line


# FuncAnimation klassen kan så lave animationen
anim = FuncAnimation(fig,  # Figuren animationen skal lave på
                     update,  # funktionen der opdaterer figuren hver frame
                     frames=len(t_arr),  # antal frames
                     interval=50,  # antal ms mellem hver frame
                     blit=True).to_jshtml()
# HTML funktionen fra IPython.display modulet kan afspille animation i notebooken
HTML(anim)
```
