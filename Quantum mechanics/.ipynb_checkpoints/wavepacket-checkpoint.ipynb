# Wavepacket
Denne Python-kode skal illustrere forskellige løsninger til den fri
partikel, som beskrevet i kapitel 2.4 i Griffith's "Introduction to
Quantum Mechanics".

Bemærk, vi har sat $\hbar = m = 1 $ i de aktuelle formler. Så vil
bølgetallet $k$ og impulsen $p$ have samme enheder. Vi starter med at importere de biblioteker og funktioner vi skal bruge

```python
import numpy as np  # Numeriske værktøjer
import matplotlib.pyplot as plt  # Plotte værktøjer
from matplotlib.animation import FuncAnimation  # Animationer
from IPython.display import HTML  # Bruges til at afspille animationen
```

## Start-bølgefunktion
Vi kigger på tre forskellige start-bølgefunktioner. Den første (waveform = 1) er gaussisk
$$
\Psi(x,0) \propto e^{-\left(\frac{x}{a}\right)^2},
$$
den næste (waveform = 2) har form som en kasse
$$
\Psi(x,0) \propto 
\begin{cases}
1 \text{ hvis } x<|a|\\
0 \text{ ellers}
\end{cases},
$$
og den sidste (waveform = 3) har form som en halv periode af en cosinus funktion
$$
\Psi(x,0) \propto 
\begin{cases}
\cos( \frac{\pi x}{2a}) \text{ hvis } x<|a|\\
0 \text{ ellers}
\end{cases},
$$
Desuden vil vi også gerne undersøge hvad der sker når vi ganger start-bølgefunktionen med en eksponentiel faktor $\Psi(x,0) \rightarrow e^{ik_0x}\Psi(x,0)$. I cellen nedenfor vælger vi hvilken bølgefunktion vi vil bruge og hvad $a$ og $k_0$ skal være parametre og plotter bølge-funktionen. 

```python
# Indstillinger for psi0 (som beskriver bølgefunktionen til t = 0)
a = 1.5
k0 = 0
waveform = 1  # 1 for gaussisk, 2 for kasse, 3 for cos, 4 for din egen funktion.

# Definer x punkter. Vi definere fra -50 til 50,
# hvilket er meget bredere en vores bølgefunktion
x_arr = np.linspace(-50, 50, 1000)

if waveform == 1:
    # Vi sætter psi0 lig med en gaussisk funktion.
    # Nå a bliver større bliver funktionen bredere
    # Bølgefunktionen er en kompleks funktion,
    # så vi skal huske at sætte datatypen til complex
    psi0 = np.exp(-(x_arr/a)**2, dtype=complex)
elif waveform == 2:
    # Vi sætter psi0 lig med en kasse funktion.
    # Nå a blever større bliver funktionen bredere
    psi0 = np.zeros(len(x_arr), dtype=complex)
    psi0[np.abs(x_arr) < a] = 1.
elif waveform == 3:
    # Vi sætter psi0 lig med en halv cos periode.
    # Nå a blever større bliver funktionen bredere
    psi0 = np.zeros(len(x_arr), dtype=complex)
    psi0[np.abs(x_arr) < a] = np.cos(x_arr[np.abs(x_arr) < a]*np.pi/(2*a))
elif waveform == 4:
    # Lav din egen startbølgefunktion nedenunder
    psi0 = np.zeros(len(x_arr), dtype=complex)

# Nu normerer vi
N = np.trapz(y=np.abs(psi0)**2, x=x_arr)
psi0 /= np.sqrt(N)

# Og ganger med en eksponential faktor
# Bemærk at denne linje ville give en fejl hvis ikke vi havde sat
# dtype=complex, da vi ellers ville forsøge at indsætte komplekse
# tal i et reelt array
psi0 *= np.exp(1j*k0*x_arr)

# Plot real og imaginær delen af start-bølgefunktion på en axes
# og normkvadratet på en anden
fig, axs = plt.subplots(1, 2, figsize=(18, 6))  # Lav en figur med 2 axes
axs[0].plot(x_arr, np.real(psi0), color='b', label=r'$Re(\psi)$')
axs[0].plot(x_arr, np.imag(psi0), color='r', label=r'$Im(\psi)$')
xlim = (-10*a, 10*a)
axs[0].set_xlim(*xlim)
axs[0].set_xlabel('$x$')
axs[0].set_ylabel(r'$\psi$')
axs[0].legend()
axs[1].plot(x_arr, np.abs(psi0)**2, color='k', label=r'$|\psi|^2$')
axs[1].set_xlim(*xlim)
axs[1].set_xlabel('$x$')
axs[1].set_ylabel(r'$|\psi|^2$')
plt.show()
```

## Start-bølgefunktion i impulsrum
Bølgefunktionen til alle tider kan bestemmes ud fra ligning (2.101), men man skal først kende $\phi(k)$. I særtilfældet $t=0$ gælder ligning (2.102):
$$
\Psi(x,0) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty\phi(k)e^{ikx}dk.
$$
Ved at gøre brug af en Fourier-transformation af dette, finder vi i ligning (2.104):
$$
\phi(k) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty\Psi(0,x)e^{-ikx}dx.
$$
Hvis man løser dette integral for de tre forskellige kurver får man 
$$
\phi(k) = \propto e^{-\left(\frac{a(k-k_0)}{2}\right)^2},
$$
for den gaussiske tilfælde (waveform = 1),
$$
\phi(k) \propto \frac{\sin(a(k-k_0))}{k-k_0},
$$
for kassen (waveform = 2) og 
$$
\phi(k) \propto \frac{\sin(a(k-k_0 + \frac{\pi}{2a}))}{k-k_0+\frac{\pi}{2a}}+\frac{\sin(a(k-k_0-\frac{\pi}{2a}))}{k-k_0-\frac{\pi}{2a}},
$$
for cosinus-funktionen (waveform = 3). 

Nedenfor regner vi $\phi(k)$ numerisk og plotter den sammen med den teoretiske kurve.

```python
# Først defineres afstanden mellem vores k punkter.
# Det er bedst hvis dk*(x_arr[-1]-x[0])=2*pi. (Tænk DFT fra fourieranalyse).
dk = 2 * np.pi / (x_arr[-1]-x_arr[0])
k_arr = (np.arange(len(x_arr)) - len(x_arr)/2)*dk  # k punkterne

# Nu kan vi regne phi(k) for alle k i k_arr ved at bruge definitionen ovenfor.
phi = [np.trapz(np.exp(-1j * k * x_arr) * psi0, x=x_arr) for k in k_arr]
# Lav om til et numpy array og skaler
phi = np.array(phi, dtype=complex) / np.sqrt(2 * np.pi)

# Nedenunder regner vi den teoretiske phi(k) kurve
phi_theo = np.zeros(len(k_arr), dtype=complex)
if waveform == 1:
    plot_theo = True    
    phi_theo = np.exp(-(a*(k_arr-k0)/2)**2)
elif waveform == 2:
    plot_theo = True    
    # sinc(x)=sin(pi*x)/(pi*x) i numpy
    phi_theo = np.sinc((k_arr-k0)*a/np.pi)
elif waveform == 3:
    plot_theo = True    
    k1 = k_arr-k0+np.pi/(2*a)
    k2 = k_arr-k0-np.pi/(2*a)
    phi_theo = np.sinc(a*k1/np.pi)+np.sinc(a*k2/np.pi)
elif waveform == 4:
    plot_theo = False    
    
    
if plot_theo:
    # Nu normerer vi
    N = np.trapz(y=np.abs(phi_theo)**2, x=k_arr)
    phi_theo /= np.sqrt(N)

fig, axs = plt.subplots(1, 2, figsize=(18, 6))
axs[0].plot(k_arr, np.real(phi), color='b', label=r'$Re(\phi)$')
axs[0].plot(k_arr, np.imag(phi), color='r', label=r'$Im(\phi)$')
if plot_theo:
    axs[0].plot(k_arr, np.real(phi_theo), 'g--',  label=r'$Re(\phi)$ theo')
    axs[0].plot(k_arr, np.imag(phi_theo), 'y--',  label=r'$Im(\phi)$ theo')
klim = (-20/a, 20/a)
axs[0].set_xlim(*klim)
axs[0].set_xlabel('$k$')
axs[0].set_ylabel(r'$\phi$')
axs[0].legend()
axs[1].plot(k_arr, np.abs(phi)**2, color='k', label=r'$|\phi|^2$')
axs[1].set_xlim(*klim)
axs[1].set_xlabel('$k$')
axs[1].set_ylabel(r'$|\phi|^2$')
plt.show()
```

## Usikkerhedsrelationen
Vi kan nu regne usikkerheden af $x$ og $k$ og sikre at usikkerhedsrelationen (husk, at $\hbar = 1$)
$$
\sigma_x\sigma_k \ge \frac{1}{2},
$$
hvor $\sigma_x = \sqrt{\text{Var}(x)} = \sqrt{\langle x^2\rangle-\langle x\rangle^2}$, og tilsvarende for $\sigma_k$. Overvej dette!!! Hvad er det, som beregnes nedenfor for $k$, og hvorfor virker det??? Et fyldestgørende svar kommer først i kapitel 3, men din intuition er måske allerede godt på vej...!

```python
meanx = np.trapz(np.abs(psi0)**2 * x_arr, x=x_arr)
meanx_sq = np.trapz(np.abs(psi0)**2 * x_arr**2, x=x_arr)
var_x = meanx_sq-meanx**2
meank = np.trapz(np.abs(phi)**2 * k_arr, x=k_arr)
meank_sq = np.trapz(np.abs(phi)**2 * k_arr**2, x=k_arr)
var_k = meank_sq-meank**2
print('Var(x)={}\nVar(p)={}\nsqrt(Var(x)Var(p))={}'.
      format(var_x, var_k, np.sqrt(var_k * var_x)))
```

## Tidsudvikling
Bølgefunktionen til et senere tidspunkt $t$ kan så findes fra ligning (2.101, med $\hbar = m = 1$):
$$
\Psi(x,t) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty\phi(k)e^{i\left(kx-\frac{k^2t}{2}\right)}dk
$$

```python
# De tider hvor bølgefunktionen bestemmes
t_arr = np.linspace(0, 3, 50)

# Gør figuren klar
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
re_line, = axs[0].plot([], [], color='b')
im_line, = axs[0].plot([], [], color='r')
axs[0].set_xlim(*xlim)
axs[0].set_ylim(-np.max(np.abs(psi0))*1.5, np.max(np.abs(psi0))*1.5)
axs[0].set_xlabel('$x$')
axs[0].set_ylabel(r'$\phi$')
abs_line, = axs[1].plot([], [], color='k')
axs[1].set_xlim(*xlim)
axs[1].set_ylim(-np.max(np.abs(psi0)**2)*1.5, np.max(np.abs(psi0)**2)*1.5)
axs[1].set_xlabel('$x$')
axs[1].set_ylabel(r'$|\phi|^2$')


# Dette er funktionen der opdaterer figuren hver frame.
def update(i):
    t = t_arr[i]
    psi_t = [np.trapz(phi*np.exp(1j*(k_arr*x-.5*k_arr**2*t)), k_arr)
             for x in x_arr]
    # Lav om til et numpy array og skaler
    psi_t = np.array(psi_t, dtype=complex) / np.sqrt(2 * np.pi)
    re_line.set_data(x_arr, np.real(psi_t))
    im_line.set_data(x_arr, np.imag(psi_t))
    abs_line.set_data(x_arr, np.abs(psi_t)**2)
    return re_line, im_line, abs_line


# FuncAnimation klassen kan så lave animationen
anim = FuncAnimation(fig,  # Figuren animationen skal lave på
                     update,  # funktionen der opdaterer figuren hver frame
                     frames=len(t_arr),  # antal frames
                     interval=50,  # antal ms mellem hver frame
                     blit=True).to_jshtml()
# HTML funktionen fra IPython.display modulet kan afspille animation i notebooken
HTML(anim)
```
