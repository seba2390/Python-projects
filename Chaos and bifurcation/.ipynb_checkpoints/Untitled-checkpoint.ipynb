```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
init_pop_size = 0.5
init_growth_fac = 0.1
max_growth = 4.0
nr_steps = int(1e4)
step_size = float(max_growth/nr_steps)

def pop_func(xn,r):
    return r*xn*(1-xn)

pop_size = init_pop_size 
growth_fac = init_growth_fac

growth_vals = []
pop_eq_vals = []
for i in range(nr_steps):
    for j in range(int(1e3)):
        pop_size = pop_func(pop_size,growth_fac)
    pop_eq_vals.append(pop_size)
    growth_vals.append(growth_fac)
    growth_fac += step_size
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(growth_vals,pop_eq_vals,'bo',ms=2)
ax.grid()
```
